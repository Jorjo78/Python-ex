```python
import pandas as pd
file_path = '/Users/giorgiocavallo/Desktop/Python/coursera/week_3/customers-100.csv'
df = pd.read_csv(file_path, sep=';', quotechar='"', skipinitialspace=True)
df.head(2)
```

## Boolean mask

```python
mask = [False for i in range (len(df))]  # for _ in range(len(df)): This is a loop that iterates over a range from 0 to len(df) - 1. The _ is a convention used when we donâ€™t care about the loop variable.
mask[3:7] = [True] * 4   ##this is a slice of the mask list. It selects elements starting from index 3 up to (but not including) index 7.
mask
```

```python
df.loc[mask]
```

### Creating masks using comparison operators

```python
mask = df.loc[:, 'First Name'].str.len() < df.loc[:, 'Last Name'].str.len()
df[mask]
```

### Pandas boolean operators  and/or/Not 

```python
df.describe()
df.info()
```

```python
mask = (df.loc[:, 'First Name'].str.len() < 4) & (df.loc[:, 'Last Name'].str.len() > 5)
```

```python
df[mask]
```
